```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('data/survey.csv', header=0, parse_dates=[0], dayfirst=True)
```

```python
data.columns
```

```python
education = data[['Höchste abgeschlossene Ausbildung', 'Hätten Sie gerne Kinder?']]
education
```

```python
crosstab = pd.crosstab(education['Höchste abgeschlossene Ausbildung'], education['Hätten Sie gerne Kinder?'])
crosstab
```

```python
sns.heatmap(crosstab, annot=True, cmap='Blues')
plt.show()
```

```python
crosstab.plot(kind='bar')
plt.show()
```

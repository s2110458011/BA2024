```python

def plot_2_categories_scatterplot(x, y, df: pd.DataFrame) -> None:
    scale = 500*df['count'].size
    size = df['count']/df['count'].sum()*scale
    self.ax.scatter(x, y, size, data=df, zorder=2)
    
    return None

def compute_counts_of_observation(raw_data: pd.DataFrame, cat1: str, cat2: str) -> pd.DataFrame:
    df_counts = raw_data.groupby([cat1, cat2]).size().reset_index()
    df_counts.columns.values[df_counts.columns==0] = 'count'
    return df_counts
```

```python
df = survey[['Alter', 'Höchste abgeschlossene Ausbildung', 'Geschlecht']]
x = 'Alter'
y = 'Höchste abgeschlossene Ausbildung'
z = 'Geschlecht'
df_counts = df.groupby([x, y, z]).size().reset_index()
df_counts.columns.values[df_counts.columns == 0] = 'count'

df_counts
```

```python
# Compute a size variable for the markers so that they have a good size regardless
# of the total count and the number of unique values in each categorical variable
scale = 50*df_counts['count'].size
size = df_counts['count']/df_counts['count'].sum()*scale

# Create matplotlib scatter plot with additional formatting
fig, ax = plt.subplots(figsize=(8,6))
scatter = ax.scatter(x, y, size, c=z, data=df_counts, zorder=2, cmap='spectral') #, color='#2A788E'
#handles, labels = scatter.legend_elements(prop="sizes", alpha=0.6)
kw = dict(prop="sizes", color=scatter.cmap(0.4), fmt="{x}",
          func=lambda s: s/50)
legend2 = ax.legend(*scatter.legend_elements(**kw), bbox_to_anchor=(1.17, 1.0), loc="best", title="Sizes", labelspacing=2)
#ax.grid(color='grey', linestyle='--', alpha=0.4, zorder=1)
ax.tick_params(length=0)
#ax.set_frame_on(False)
#ax.margins(.1)
```
